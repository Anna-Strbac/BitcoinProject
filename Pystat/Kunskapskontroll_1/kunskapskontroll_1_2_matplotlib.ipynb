### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

There are two major ways to use Matplotlib:

1.An "object-oriented" interface. It uses methods on a Figure or Axes object to create other Artists, and build a visualization step by step.

2.An implicit "pyplot" interface. It keeps track of the last Figure and Axes created, and adds Artists to the object it thinks the user wants.

For more complex and structured plotting it is recommended to use the "object-oriented" interface. It is the most flexible and underlies the other interfaces, better for creating multi-subplot figures, and allows us to have full control over the plots. 

While the implicit "pyplot" interface is quick and convenient for simple plots, it can become limiting for more advanced plotting tasks. In implicit "pyplot", we create plots without explicitly creating figure and axes objects. Instead we use functions provided by pyplot, which internally manage these objects for us. We can think of it as a more concise way of creating simple plots without dealing with the underlying figure and axes details.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

The "figure" in Matplotlib is used to create an empty graphical window and display it on the screen. This window becomes the current figure, which means that all subsequent graphical functions will produce plots within this window. 

An Axes is an Artist attached to a Figure that contains a region for plotting data, and usually includes two (or three in the case of 3D) Axis objects. It's the area where we can set labels, titles, and customize the appearance of the plot.

Axis set the scale and limits and generate ticks (the marks on the Axis) and ticklabels (strings labeling the ticks). 

Artist is everything visible on the Figure, including figure, axes and axis. This includes Text objects, Line2D objects, collections objects, Patch objects, etc. 

### When plotting in Matplotlib, what is the expected input data type?

The expected input data type for plotting in Matplotlib is typically NumPy arrays or standard Python lists for the data to be plotted. 

Most methods will also parse a string-indexable object like a dict, a structured numpy array, or a pandas.DataFrame. 

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

The object-oriented approach:

```python
x = np.linspace(-4, 4, 100)
y = x**2

fig, ax = plt.subplots()

ax.plot(x, y)

ax.set_title("y = x^2")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

The pyplot approach:

```python
plt.plot(x, y)
plt.title("y = x^2")
plt.xlabel('x')
plt.ylabel('y')
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, axs = plt.subplots(1, 2, figsize = (10,5), layout = 'constrained')

# Scatter plot
axs[0].scatter(random_data_x, random_data_y, alpha=0.5)

axs[0].set_title('Scatter Plot')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

# Bar plot
axs[1].bar(names, values)

axs[1].set_title('Bar Plot')
axs[1].set_ylabel('Amount')
```
